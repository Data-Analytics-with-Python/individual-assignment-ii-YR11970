### Task B.1 (15 points)

i. Consider that customers are grouped by their **education levels**. Develop a bar chart with several sets of bars, in which:

* Each set of bars (i.e.,bar of the same color) represent the spendings by customers in different groups on a spefic product category. For example, spending by customers "Graduation", "Master", "PhD", ..., on *fish products*.  
* The chart should allow you to easily visuallize both the **total spending** by customers in a group on all product categories and the **composition** of their spending on those product categories.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
* Which customer group contributes the highest revenue?

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
# Spending columns
spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts',
              'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
education_col = 'Education'  # check your actual column name if needed

# 1. Melt the data for grouped bar plotting
df_long = df.melt(id_vars=[education_col], value_vars=spend_cols,
                  var_name='Product', value_name='Spend')

# 2. Grouped bar chart: total by product type per education level
fig = px.bar(
    df_long,
    x='Product',
    y='Spend',
    color=education_col,
    barmode='group',
    title='Customer Spendings by Product Category and Education Level',
    labels={'Spend': 'Total Spend', 'Product': 'Product Category', education_col: 'Education'}
)

fig.update_traces(opacity=1, marker_line_width=0)

fig.show()

# 3. To also see composition by education (optional, stacked view):
fig2 = px.bar(
    df_long.groupby([education_col, 'Product']).sum(numeric_only=True).reset_index(),
    x=education_col,
    y='Spend',
    color='Product',
    barmode='stack',
    title='Spending Composition by Education Level',
    labels={'Spend': 'Total Spend', education_col: 'Education', 'Product': 'Product Category'}
)

fig2.update_traces(opacity=1, marker_line_width=0)

fig2.show()
```

Based on the bar charts, the "Graduation" education group contributes the highest revenue among all customer groups.

In both the grouped and stacked bar charts, the total spending (the sum of all bars or the tallest stacked segment) for "Graduation" far exceeds that of other education levels in every product categoryâ€”especially in high-value categories like Wines and Meat Products.

The "Graduation" group consistently has the highest bars across nearly every category, indicating that customers with this education background collectively spend the most and are the supermarketâ€™s top revenue contributors.

## Part C (30 points)

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
# Create a 'Generation' column
def get_generation(year):
    if 1946 <= year <= 1964:
        return "Baby Boomer"
    elif 1965 <= year <= 1980:
        return "Gen X"
    elif 1981 <= year <= 1996:
        return "Millennial"
    elif 1997 <= year <= 2012:
        return "Gen Z"
    else:
        return "Other"

# Adapt column name to your actual CSV:
df['Generation'] = df['Year_Birth'].apply(get_generation)

# Focus only on Gen X and Millennials (as requested)
filtered = df[df['Generation'].isin(['Gen X','Millennial'])]

# i. Distribution by Generation (box plot)
fig = px.box(
    filtered,
    x='Generation',
    y='NumWebPurchases',
    points='all',
    title='Distribution of Online Purchases by Generation',
    labels={'NumWebPurchases': 'Number of Online Purchases'}
)
fig.show()

# iii. Further breakdown by number of kids at home (Kidhome)
fig2 = px.box(
    filtered,
    x='Generation',
    y='NumWebPurchases',
    color='Kidhome',
    points='all',
    title='Online Purchases by Generation and Number of Kids at Home',
    labels={'NumWebPurchases': 'Number of Online Purchases', 'Kidhome': 'Kids at Home'}
)
fig2.show()
```

**How would you comment on the pattern of online purchases across different generations?**

  The box plot comparing Gen X and Millennial customers shows that both generations have a similar range and distribution of online purchases, with most customers making between 1 and 6 online purchases.
  
  The median number (the line inside each box) is about 3 for Gen X and slightly above 3 for Millennials.

Gen X customers are more likely to be frequent online shoppers than Millennials, as shown by the higher density of points in the upper purchase range on the boxplot.

Gen X has a few clear outliersâ€”some customers purchase far more online than the majority, with a maximum up to 25, while Millennials' outliers are not as extreme.

Overall, both groups show considerable variability, but Gen X displays slightly more extreme high outliers.



**How would you comment on the pattern of online purchases across the different subgroups of customers with number of kids at home?**

When breaking down by number of kids at home, the spread and center of the distribution differ:

Customers with 0 kids at home have a wide range and include most of the high outliers, indicating the highest variability.

Customers with 1 kid at home generally have a median similar to those with 0 kids, though the spread and number of outliers are somewhat reduced.

Customers with 2 kids at home tend to make fewer online purchases, with a lower median and less spread, suggesting that larger families in both generations make fewer online purchases and show less variability.

The number of kids at home moderates online activity: Fewer kids is associated with higher online purchasing and more variability, while larger numbers of kids correlate with fewer and more consistent (less variable) online purchases.
